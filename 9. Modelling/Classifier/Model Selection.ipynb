```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import cohen_kappa_score
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df1 = pd.read_csv("final_train_al.csv",index_col=0)
df2 = pd.read_csv("final_train_pl.csv",index_col=0)
df3 = pd.read_csv("final_train_ll.csv",index_col=0)
```

```python
X1= df1.drop(['class'], axis=1)
Y1 = df1['class']
X2= df2.drop(['class'], axis=1)
Y2 = df2['class']
X3= df3.drop(['class'], axis=1)
Y3 = df3['class']
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size = 0.2)
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2, Y2, test_size = 0.2)
X3_train, X3_test, Y3_train, Y3_test = train_test_split(X3, Y3, test_size = 0.2)
```

```python
# define models
# Logistic Regression classifier(lambda = 0.1, 1, 10,100)
lr = LogisticRegression(penalty = 'l2', random_state = 0)

# Random Forest Classifier
rf = RandomForestClassifier(n_estimators = 90, oob_score = True, criterion = 'entropy')

# eXtreme Gradient Boosting Classifier (XGBClassifier)
xgb = XGBClassifier(max_depth=20, n_estimators=200,objective='multi:softmax',num_class=3)

# Balanced Bagging Classifier
bb = BalancedBaggingClassifier(base_estimator = RandomForestClassifier(criterion='gini'), n_estimators = 30, bootstrap = True)

# CatBoost Classifier
cc = CatBoostClassifier(iterations=1000, depth=8, learning_rate=0.2)

# Naive Bayesian
gn = GaussianNB()
```

```python
score = {'lr':[],'rf':[],'xgb':[],'bb':[],'cc':[],'gn':[]}
for data in [df1,df2,df3]:
    X1= data.drop(['class'], axis=1)
    Y1 = data['class']
    X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size = 0.2)
    lr.fit(X1_train, Y1_train)
    y_predict = lr.predict(X1_test)
    score['lr'].append(cohen_kappa_score(Y1_test,y_predict))
    rf.fit(X1_train, Y1_train)
    y_predict = rf.predict(X1_test)
    score['rf'].append(cohen_kappa_score(Y1_test,y_predict))
    xgb.fit(X1_train, Y1_train)
    y_predict = xgb.predict(X1_test)
    score['xgb'].append(cohen_kappa_score(Y1_test,y_predict))
    bb.fit(X1_train, Y1_train)
    y_predict = bb.predict(X1_test)
    score['bb'].append(cohen_kappa_score(Y1_test,y_predict))
    cc.fit(X1_train, Y1_train)
    y_predict = cc.predict(X1_test)
    score['cc'].append(cohen_kappa_score(Y1_test,y_predict))
    gn.fit(X1_train, Y1_train)
    y_predict = gn.predict(X1_test)
    score['gn'].append(cohen_kappa_score(Y1_test,y_predict))
```

```python
score
```

```python
# improve xgboost
score_dict = {}
dep_list = np.arange(16,26,2)
n_estimators_list = np.arange(160,260,20)
idx = 0
for data in [df1,df2,df3]:
    idx += 1
    for dep in dep_list:
        for es in n_estimators_list:
            X1= data.drop(['class'], axis=1)
            Y1 = data['class']
            X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size = 0.2)
            xgb = XGBClassifier(max_depth=dep, n_estimators=es,objective='multi:softmax',num_class=3)
            xgb.fit(X1_train, Y1_train)
            y_predict = xgb.predict(X1_test)
            score_dict[len(score_dict)] = {'dep':dep, 'es':es, 'score':cohen_kappa_score(Y1_test,y_predict)}
```

```python
score_df = pd.DataFrame(data = score_dict).T
```

```python
score_df['data'] = ['df1','df2','df3']*25
```

```python
for i in range(len(score_df)):
    if score_df['score'][i] == max(score_df['score']):
        print(score_df.iloc[i])
```
